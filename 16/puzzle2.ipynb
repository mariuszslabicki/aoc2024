```python
import copy

from collections import deque

tiles = set()
start = None
end = None
dir = (1, 0)

max_x = None
max_y = None

with open('input') as f:
    for y, line in enumerate(f):
        max_y = y
        for x, char in enumerate(line.strip()):
            if char != '#':
                tiles.add((x, y))
            if char == 'S':
                start = (x, y)
            if char == 'E':
                end = (x, y)
            max_x = x

path = deque()
path.append(start)

paths = [[0, dir, path]]

debug_counter = 0

found = False

min_len = (max_x * max_y)**2

visited = dict()

while True:
    new_paths = []
    if len(paths) == 0:
        break
    for path in paths:
        steps = [(0, 1), (0, -1), (1, 0), (-1, 0)]
        for step in steps:
            candidate = (path[2][-1][0]+step[0], path[2][-1][1]+step[1])
            if candidate not in path[2] and candidate in tiles:
                new_path = [path[0], path[1], []]
                if new_path[1] == step:
                    new_path[0] += 1
                else:
                    new_path[0] += 1001
                new_path[1] = step
                if new_path[0] > min_len:
                    continue
                if (candidate, new_path[1]) in visited and visited[(candidate, new_path[1])] < new_path[0]:
                    continue
                visited[(candidate, new_path[1])] = new_path[0]

                new_path[2] = path[2].copy() 
                new_path[2].append(candidate)

                if new_path[2][-1] == end:
                    found = True
                    
                    if path[0] + 1 < min_len:
                        min_len = path[0] + 1
                    continue
                new_paths.append(new_path)
    paths = new_paths

## Part 2

path = deque()
path.append(start)

paths = [[0, dir, path]]

visited = dict()

while True:
    new_paths = []

    for path in paths:
        steps = [(0, 1), (0, -1), (1, 0), (-1, 0)]
        for step in steps:
            candidate = (path[2][-1][0]+step[0], path[2][-1][1]+step[1])
            if candidate not in path[2] and candidate in tiles:
                new_path = [path[0], path[1], []]
                if new_path[1] == step:
                    new_path[0] += 1
                else:
                    new_path[0] += 1001
                new_path[1] = step
                if new_path[0] > min_len or new_path[0] == min_len and candidate != end:
                    continue
                if (candidate, new_path[1]) in visited and visited[(candidate, new_path[1])] < new_path[0]:
                    continue
                visited[(candidate, new_path[1])] = new_path[0]

                new_path[2] = path[2].copy() 
                new_path[2].append(candidate)

                new_paths.append(new_path)

    if len(new_paths) == 0:
        break
    paths = new_paths

seatings = set()
seatings.update([start, end])

for path in paths:
    if path[2][-1] == end:
        seatings.update(path[2])
        
print(len(seatings))
```
