```python
positions = []
vectors = []

max_x = 101
max_y = 103

with open('input') as f:
    game = []
    for y, line in enumerate(f):
        line = line.split()
        x = int(line[0][2:].split(",")[0])
        y = int(line[0][2:].split(",")[1])
        px = int(line[1][2:].split(",")[0])
        py = int(line[1][2:].split(",")[1])
        positions.append([x, y])
        vectors.append([px, py])
time = 0
while True:       
    left_sides = [0 for i in range(max_y)]
    right_sides = [0 for i in range(max_y)]

    for i in range(len(positions)):
        positions[i][0] = (positions[i][0] + vectors[i][0]) % max_x
        positions[i][1] = (positions[i][1] + vectors[i][1]) % max_y

    if time % 101 == 11 and time % 103 == 87:
        print("Time", time+1)
        for y in range(max_y):
            for x in range(max_x):
                if ([x, y]) in positions:
                    print("*", end="")
                else:
                    print(" ", end="")
            print("")
        break
    time += 1
    
```
