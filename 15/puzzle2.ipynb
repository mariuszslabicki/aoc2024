```python
import copy

walls = set()
boxes = set()
steps = []

pos = None

reading_steps = False

with open('input') as f:
    for y, line in enumerate(f):
        if len(line) == 1:
            reading_steps = True
        for x, char in enumerate(line.strip()):
            if reading_steps is False:
                if char == '.':
                    continue
                if char == '#':
                    walls.add((2*x, y))
                if char == 'O':
                    boxes.add((2*x, y))
                if char == '@':
                    pos = (2*x, y)
                max_x = x
                max_y = y
            else:
                steps.append(char)

for step in steps:
    if step in ['>', '<']:
        if step == '>':
            shift = (1, 0)
            jump = (2, 0)
            next_pos = (pos[0] + 1, pos[1])
            pointer = (pos[0] + 1, pos[1])
        if step == '<':
            shift = (-1, 0)
            jump = (-2, 0)
            next_pos = (pos[0] - 1, pos[1])
            pointer = (pos[0] - 2, pos[1])

        if pointer in walls:
            continue
        if pointer in boxes:
            to_remove = set()
            while pointer in boxes:
                to_remove.add(pointer)
                pointer = (pointer[0] + jump[0], pointer[1])
            if pointer in walls:
                continue
            for element in to_remove:
                boxes.remove(element)
                boxes.add((element[0]+shift[0], element[1]))
            pos = next_pos
            continue
        pos = next_pos
        continue

    if step in ['v', '^']:
        if step == 'v':
            shift = (0, 1)
        if step == '^':
            shift = (0, -1)

        next_pos = (pos[0] + shift[0], pos[1] + shift[1])
        next_pos_left = (pos[0] + shift[0] - 1, pos[1] + shift[1])

        if next_pos in walls or next_pos_left in walls:
            continue

        if next_pos not in walls and next_pos not in boxes and next_pos_left not in walls and next_pos_left not in boxes:
            pos = next_pos
            continue

        if next_pos in boxes:
            pointer = next_pos
        if next_pos_left in boxes:
            pointer = next_pos_left
        row = set()
        row.add(pointer)
        to_remove = set()
        to_remove.add(pointer)

        wall = False

        while len(row) > 0:
            new_row = set()
            for element in row:
                if (element[0]-1, element[1] + shift[1]) in walls:
                    wall = True
                    break
                if (element[0]-1, element[1] + shift[1]) in boxes:
                    new_row.add((element[0]-1, element[1] + shift[1]))
                    to_remove.add((element[0]-1, element[1] + shift[1]))

                if (element[0], element[1] + shift[1]) in walls:
                    wall = True
                    break
                if (element[0], element[1] + shift[1]) in boxes:
                    new_row.add((element[0], element[1] + shift[1]))
                    to_remove.add((element[0], element[1] + shift[1]))

                if (element[0]+1, element[1] + shift[1]) in walls:
                    wall = True
                    break
                if (element[0]+1, element[1] + shift[1]) in boxes:
                    new_row.add((element[0]+1, element[1] + shift[1]))
                    to_remove.add((element[0]+1, element[1] + shift[1]))
                row = copy.deepcopy(new_row)
                to_remove.update(copy.deepcopy(new_row))
            if wall is True:
                break
        if wall is True:
            continue
        for element in to_remove:
            boxes.remove(element)
        for element in to_remove:
            boxes.add((element[0] + shift[0], element[1] + shift[1]))
            pos = next_pos

total_coordinates = 0

for box in boxes:
    total_coordinates += box[1] * 100 + box[0]

print(total_coordinates)
```
