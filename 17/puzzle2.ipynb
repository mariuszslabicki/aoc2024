```python
A = None
B = None
C = None
program = None
with open('input') as f:
    for y, line in enumerate(f):
        line = line.strip().split()
        if len(line) == 3:
            if line[1] == 'A:':
                A = int(line[2])
            if line[1] == 'B:':
                B = int(line[2])
            if line[1] == 'C:':
                C = int(line[2])
        if len(line) == 2:
            program = line[1].split(",")
            program = [int(x) for x in program]

def evaluate(program, A, B, C):
    ip = 0
    output = []
    while ip < len(program):
        opcode = program[ip]
        literal_operand = program[ip+1]
        if literal_operand in [0, 1, 2, 3]:
            combo_operand = literal_operand
        elif literal_operand == 4:
            combo_operand = A
        elif literal_operand == 5:
            combo_operand = B
        elif literal_operand == 6:
            combo_operand = C
        # elif literal_operand == 7:
        #     print("Error, reserved operand")
        #     break
        if opcode == 0:
            A = A // 2**combo_operand
            ip += 2
        elif opcode == 1:
            B = B ^ literal_operand
            ip += 2
        elif opcode == 2:
            B = combo_operand % 8
            ip += 2
        elif opcode == 3:
            if A == 0:
                ip += 2
                continue
            else:
                ip = literal_operand
        elif opcode == 4:
            B = B ^ C
            ip += 2
        elif opcode == 5:
            out_val = combo_operand%8
            output.append(out_val)
            ip += 2
        elif opcode == 6:
            B = A // 2**combo_operand
            ip += 2
        elif opcode == 7:
            C = A // 2**combo_operand
            ip += 2
    return output

A = 0

while True:
    output = evaluate(program, A, B, C)
    output_len = len(output)

    if program[-len(output):] == output:
        if output_len == len(program):
            print(A)
            break
        A = 8**(output_len)
        continue
    
    diff_exp = 0
    for i in range(len(output)-1):
        output_el = output[-i-1]
        program_el = program[-i-1]
        if output_el != program_el:
            diff_exp = len(output) - 1 - i
            break

    A += 8**diff_exp
```
